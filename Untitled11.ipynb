# Задание 2

Теперь напишем цикл, который найдет категорию людей, у которых к 30 дефолтность не ниже 25%.

```python
ans = []
for act in odobr['ActivityType'].unique():
    for edu in odobr['Education'].unique():
        for purp in odobr['LoanPurpose'].unique():
            filtered_df = odobr[(odobr['ActivityType'] == act) & 
                               (odobr['Education'] == edu) & 
                               (odobr['LoanPurpose'] == purp)]
            count_zero_summ = len(filtered_df[filtered_df['summ'] == 0])
            total_count = len(filtered_df)
            
            if total_count != 0:
                if count_zero_summ / total_count > 0.25 and total_count > 1000:
                    ans.append([act, edu, purp])
```

Напишем функцию upsample, которая будет выравнивать дисбаланс классов, если он будет.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

#features, target = upsample(features, target, 7)
```

Видим метрики обучения, напишем функцию метрики gini

```python
import numpy as np

def gini_coefficient(income, population):
    income = np.sort(income)
    cum_population = np.cumsum(np.sort(population))
    cum_income = np.cumsum(income * population)
    
    total_income = cum_income[-1]
    total_population = int(cum_population[-1])  # Преобразуем в целое число
    
    perfect_income_distribution = np.linspace(0, total_income, total_population)
    area_below_curve = np.trapz(perfect_income_distribution) - np.trapz(cum_income)
    area_total = np.trapz(perfect_income_distribution)
    
    gini_coefficient = area_below_curve / area_total
    return gini_coefficient

gini_coefficient(model.predict(features_test), target_test.tolist())
```

Чем ниже значение метрики gini, тем лучше. Видим, что наша модель вполне осмысленна. Посмотрим на другие метрики

Видим, что есть веса, которые вообще не влияют на результат, например, 9.64510965e-04. Построим модель XGBoost, случайный лес с градиентным бустингом, подберем параметры

```python
depth_1 = 0
n1 = 0
roc = 0
model_xb = None
for depth in range(1, 10):
    for n in range(1,10):
        xb = xgboost.XGBClassifier(n_estimators = n, max_depth = depth, random_state = 123, iterations = 150, learning_rate = 0.0001)
        xb.fit(features_train, target_train)
        probabilities_valid = xb.predict_proba(features_test)
        probabilities_one_valid = probabilities_valid[:,1]
        if roc < roc_auc_score(target_test, probabilities_one_valid):
            roc = roc_auc_score(target_test, probabilities_one_valid)
            n1 = n
            model_xb = xb
            depth_1 = depth
print('roc_auc = ', roc)
```
